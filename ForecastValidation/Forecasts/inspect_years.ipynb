```python
NIIEEs = [data_2016['NIIEE_2'].mean(), data_2017['NIIEE_2'].mean(), data_2018['NIIEE_2'].mean(), data_2019['NIIEE_2'].mean()]
val_loss = [0.42737, 0.34638, 0.35348, 0.35270]
years = ['2016', '2017', '2018', '2019']
```

```python
plt.rcParams.update({'font.size': 12})

fig = plt.figure(figsize = (7.5, 7.5))
ax = fig.add_subplot()
lns1 = ax.plot(NIIEEs, '--o', label = 'NIIEE')
ax.grid()

ax2 = ax.twinx()

lns2 = ax2.plot(val_loss, '--o', c='tab:orange', label = 'Val_loss')
ax2.set_ylim([0.34, 0.45])
ax2.set_ylabel('Validation loss')

ax.set_xticks(np.arange(4), years)
ax.set_title('Forecast error with increasing training data')
ax.set_ylabel('Mean annual ice edge displacement error [km]')
ax.set_xlabel('First year of training data')

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)

plt.savefig('/lustre/storeB/users/arefk/MScThesis_AreKvanum2022_SeaIceML/ForecastValidation/Forecasts/years_start.pdf')
```
